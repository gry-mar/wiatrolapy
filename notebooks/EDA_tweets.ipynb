## Preprocessing tekstu

```python
import re


# Remove punctuation
tweets_all["text"] = tweets_all["text"].apply(
    lambda x: re.sub("[,\.!?]", "", x) if not pd.isna(x) else ""
)
tweets_all["text"] = tweets_all["text"].apply(
    lambda x: re.sub(r"https?:\/\/.*\s", "", x) if not pd.isna(x) else ""
)


# Convert the titles to lowercase
tweets_all["text"] = tweets_all["text"].map(
    lambda x: x.lower() if not pd.isna(x) else ""
)
# # Print out the first rows of papers
tweets_all["text"].iloc[2]
```

```python
def remove_tags_and_hashtags(text):
    # Remove hashtags and @ tags
    text = re.sub(r'#', '', text)
    text = re.sub(r'@', '', text)
    return text.strip()

tweets_all["text"] = tweets_all["text"].apply(remove_tags_and_hashtags)
tweets_all
```

```python
import nltk
import spacy
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import SnowballStemmer
from spacy.lang.pl import Polish
import re

# Download necessary resources for NLTK
nltk.download('punkt')
with open('data/polish_stopwords.txt', 'r', encoding="utf-8") as f:
        stopwords = f.readlines()
        stopwords_strip = set(m.strip() for m in stopwords)
        stopwords_set = list(frozenset(stopwords_strip))

stopwords_set.extend(['Tę', 'tę', 'https', 'nan', 'link', "\\'\\",
                    "-|", "\\'\\'", "+/-", "``", "\\-", "\'\'",
                        "|0, ", "”", "( ͡° ͜ʖ ͡°)", "...", "@", "#" ])

# Function to preprocess text
def preprocess_text(text):
    # Tokenization
    tokens = word_tokenize(text.lower())  # Convert to lowercase for consistency

    # Remove stopwords
    stop_words = stopwords_set
    tokens = [token for token in tokens if token not in stop_words]

    # Lemmatization (using spaCy for Polish)
    nlp = Polish()
    doc = nlp(' '.join(tokens))
    lemmatized_tokens = [token.lemma_ for token in doc]
    

    # Join the tokens back into a string
    preprocessed_text = ' '.join(lemmatized_tokens)

    return preprocessed_text

# Apply the preprocessing function to the 'text' column
tweets_all['processed_text'] = tweets_all['text'].apply(preprocess_text)

# Display the DataFrame with the processed text
tweets_all
```

### Liczba słów i tokenów

```python
def count_unique_tokens(df):

    tokens = []
    for tok in df["tokens"]:
        tokens.extend(tok)

    unique_tokens = set(tokens)
    num_unique_tokens = len(unique_tokens)
    return num_unique_tokens
```

### Liczba występień części mowy

```python
def count_pos(text):
    doc = nlp(text)
    pos_counts = {}
    for token in doc:
        pos = token.pos_
        if pos in pos_counts:
            pos_counts[pos] += 1
        else:
            pos_counts[pos] = 1
    return pos_counts
```

```python
tweets_all["pos_counts"] = tweets_all["text"].progress_apply(count_pos)
```

```python
pos_df = tweets_all['pos_counts'].apply(pd.Series)
tweets_all = pd.concat([tweets_all, pos_df], axis=1)
tweets_all = tweets_all.drop('pos_counts', axis=1)
tweets_all[tweets_all.columns[7:]] = tweets_all[tweets_all.columns[7:]].fillna(0)
```

```python
dictionary = Dictionary(tweets_all['tokens'])
corpus = [dictionary.doc2bow(doc) for doc in tweets_all['tokens']]

# Build an LDA model
lda_model = LdaModel(corpus, num_topics=6, id2word=dictionary, passes=10)

# Assign topics to documents and add a 'topic' column to the DataFrame
tweets_all['topic'] = [max(lda_model[doc], key=lambda x: x[1])[0] for doc in corpus]
```

```python
from textblob import TextBlob
def sentiment_analysis():
 def getSubjectivity(text):
   return TextBlob(text).sentiment.subjectivity
  
 #Create a function to get the polarity
 def getPolarity(text):
   return TextBlob(text).sentiment.polarity
  
 #Create two new columns ‘Subjectivity’ & ‘Polarity’
 tweets_all['TextBlob_Subjectivity'] = tweets_all['text'].apply(getSubjectivity)
 tweets_all['TextBlob_Polarity'] = tweets_all['text'].apply(getPolarity)

sentiment_analysis()
tweets_all
```

```python
from tqdm import tqdm
import math
import numpy as np
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

RECREATE_SENTIMENT = True

if RECREATE_SENTIMENT:
    with torch.no_grad():
        # id2label = {0: "negative", 1: "neutral", 2: "positive"}
        mps_device = torch.device("mps")
        id2label = {0: -1, 1: 0, 2: 1}
        tokenizer = AutoTokenizer.from_pretrained(
            "Voicelab/herbert-base-cased-sentiment"
        )
        model = AutoModelForSequenceClassification.from_pretrained(
            "Voicelab/herbert-base-cased-sentiment"
        ).to(mps_device)
        tweets_all['sentiment_herbert'] = np.NaN
        sentiment_list = []
        for i in tqdm(range(len(tweets_all))):
            input = tweets_all['text'].iloc[i]
            encoding = tokenizer(
            input,
            add_special_tokens=True,
            return_token_type_ids=True,
            truncation=True,
            padding='max_length',
            return_attention_mask=True,
            return_tensors='pt',
            ).to("mps")
            output = model(**encoding).logits.to("cpu").detach().numpy()
            prediction = id2label[np.argmax(output)]
            sentiment_list.append(prediction)
```

```python
tweets_all['sentiment_herbert'] = sentiment_list
tweets_all
```
